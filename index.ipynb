Start by investigating the database stored in lego.db and joining the tables into a unified dataset!

> **Hint:** use this SQL query to preview the tables in an unknown database:
```sql
SELECT name FROM sqlite_master
            WHERE type='table'
            ORDER BY name;
```

````python
import sqlite3
import pandas as pd
conn = sqlite3.connect('lego.db')
cur = conn.cursor()
cur.execute('''
            SELECT name 
            FROM sqlite_master
                WHERE type='table'
                ORDER BY name;
            '''
           )
schema = pd.DataFrame(cur.fetchall())
schema[0]
````

````python
for x in schema[0]:
    cur.execute(
        f'''
        SELECT *
        FROM {x};
        '''
    )

    df = pd.DataFrame(data = cur.fetchall())
    df.columns = [i[0] for i in cur.description]
    display(df.head())
````

````python
cur.execute(
    f'''
    SELECT *
    FROM {schema[0][0]}
    JOIN {schema[0][1]}
    USING (prod_id)
    JOIN {schema[0][2]}
    USING (prod_id)
    JOIN {schema[0][3]}
    USING (prod_id);
    '''
)

df = pd.DataFrame(data = cur.fetchall())
df.columns = [i[0] for i in cur.description]
display(df.head(), df.shape)
````
